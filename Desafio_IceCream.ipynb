```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# LEndo a base de dados
icecream = pd.read_csv('IceCreamData.csv')
```

```python
#Ajustando os dados para valores inteiros

# Converter para inteiros
icecream['Revenue'] = icecream['Revenue'].astype(int)

```

```python
#Limparndo o dataFrame
icecream = icecream.dropna()

```

```python
#Convertendo de graus celcius para Fahrenheit
icecream['Temperature'] = (icecream['Temperature'] * 9/5) + 32

#Convertendo a temperatura para inteiro
icecream['Temperature'] = icecream['Temperature'].astype(int)
```

```python
# filtradno o dataframe para que tenhaapenas temperaturas maior que 40 e menor que 80 Fahrenheit
icecream = icecream.loc[(icecream['Temperature'] >= 40) & (icecream['Temperature'] <= 80) & (icecream['Revenue'] >= 250) & (icecream['Revenue'] <= 550)]
```

```python
# limpadno a base para pegar apenas os multiplos de 10
icecream = icecream.loc[(icecream['Temperature'] % 10 == 0) & (icecream['Temperature'] <= 80) & (icecream['Revenue'] >= 250) & (icecream['Revenue'] <= 550)]
```

```python
# pegando apenas uma certa quantidade de linhas do dataframe aleatoriamente

icecream = icecream.sample(12)
icecream
```

```python
plt.bar(icecream['Temperature'], icecream['Revenue'], color='red', label='US$')

plt.title('Ice Cream Truck')
# plt.yticks(range(250, 551, 50))
plt.xticks(range(50, 71, 5))
plt.ylabel('Scales ( US$)')
plt.legend()
plt.xlabel('Temperatura (Fahrenheit)')
plt.show()
```
