```python
#Import das bibliotecas
import pandas as pd
```

```python
#Acessadno o arquivo csv
titanic = pd.read_csv('titanic.csv')
titanic
```

```python
# Fazendo um filtro para selecionar apenas algumas colunas
# Minha solução
titanic = titanic[['Name', 'Age', 'Sex', 'Pclass', 'Survived']]
titanic

# Solução professor
filter1 = titanic[['Name', 'Age', 'Sex', 'Pclass', 'Survived']]
filter1
```

```python
#Fazendo um filtro para pegar apenas o sexo feminino

# MInha solução
titanic = titanic[titanic['Sex'] == 'female']
titanic

# Solução professor
filter1 = filter1.loc[(filter1['Sex'] == 'female') & (filter1['Pclass'] == 1) & (filter1['Survived'] == 1)]
filter1
```

```python
#Filtro para pegar apenas a primeira classe
titanic = titanic[titanic['Pclass'] == 1]
titanic
```

```python
#Filtro para pegar apenas os sobreviventes
titanic = titanic[titanic['Survived'] == 1]
titanic
```

```python
# Deixando em ordem alfabetica
#Minha solução
titanic = titanic.sort_values(by='Name')
titanic

#Solução professor
filter1 = filter1.sort_values(by='Name')
filter1
```

```python
# Salvando em um novo arquivo csv
titanic.to_csv('titanic_result.csv', index=False)
```
